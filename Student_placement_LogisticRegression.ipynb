```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

```python
df = pd.read_csv('/content/Placement_Data_Full_Class.csv')
df.head()
```

```python
df.drop(labels=['sl_no','salary'],axis = 1 , inplace=True)
df.head()
```

```python
temp = pd.get_dummies(df['gender']).drop(labels='F' , axis = 1)
```

```python
df.drop(labels='gender' , axis = 1 , inplace = True)
df = pd.concat([df,temp] , axis = 1)
df.head()
```

```python
temp1 = pd.get_dummies(data=df).drop(labels=['ssc_b_Others' , 'hsc_b_Others','hsc_s_Arts','degree_t_Others','workex_No','specialisation_Mkt&HR','status_Not Placed'] , axis =1)
temp1.head()
```

```python
temp1.iloc[:,0:5] = preprocessing.StandardScaler().fit_transform(temp1.iloc[:,0:5])
temp1.head()
```

```python
temp1['status_Placed'].value_counts()
```

```python
oversample = SMOTE()
X, Y = oversample.fit_resample(temp1.iloc[:,:-1], temp1.loc[:,['status_Placed']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
clf = LogisticRegression().fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
